```python
df=pd.read_csv("/kaggle/input/london-bike-sharing-dataset/london_merged.csv")
df.head()
```

```python
new_columns={'timestamp':'Time',"cnt":'Count',"t1":"Actual_Temp","t2":"Temp_feels_like",'hum':"humidity_percentage","wind_speed":"wind_speed_kph",'weather_code':"weather","is_holiday":"is_holiday",'is_weekend':'is_weekend',"season":'season' }
```

```python
df.rename(new_columns,axis=1,inplace=True)
```

```python
df.head()
```

```python
df.humidity_percentage=df.humidity_percentage/100
```

```python
df["weather"].nunique
```

```python
df['weather'].value_counts()
```

```python
weather_col={'1.0':"clear",
            '2.0':"Scattered clouds",
            "3.0":"Broken clouds",
            "4.0":"Cloudy",
            "7,0":'Rain',
            "10.0":"Rain with thunderstorm",
            "26.0":"Snowfall"}

```

```python
df.weather=df.weather.astype('str')
df.weather=df.weather.map(weather_col)
```

```python
df.head()
```

```python
df['season'].value_counts()
```

```python
season_col={'0.0':"spring",
           "1.0":"summer",
           "2.0":"autumn",
           "3.0":"winter"}
df.season=df.season.astype('str')
df.season=df.season.map(season_col)
```

```python
df['season']
```

```python
df.to_excel("london_bike.xlsx",sheet_name='data')
```
